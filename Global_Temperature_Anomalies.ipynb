```python
df= df.dropna(axis=1)
```

```python
df= df.drop(columns=['J-D', 'MAM', 'JJA', 'SON'])
```

```python
df = df[df['Hemisphere']=='Global'] #condition
```

```python
df
```

```python
df.columns
```

```python
df['Yearly Mean'] = df[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul',
       'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].mean(axis=1) #creating a new column with the yearly mean
```

```python
df
```

```python
df1 = df[['Yearly Mean','Year']].copy() #creating a copy of the df for the viz 
```

```python
df1['Yearly Average above/below baseline Â°C']= df1['Yearly Mean'].round(decimals=3)
```

```python
import plotly.express as px
fig = px.bar_polar(df1, r='Yearly Average above/below baseline Â°C', template="plotly_dark",color='Yearly Average above/below baseline Â°C',hover_name='Year',
                   color_discrete_sequence= px.colors.sequential.Plasma_r,start_angle = 90, log_r = False,title ='Global temperature anomalies compared to baseline, from 1880 to 2020',width=1620, height=800)
fig.show()
```

```python
fig.update_layout(
    polar = dict(
      angularaxis = dict(
            tickvals=[7, 45, 90, 135, 180, 225, 270, 315, 356],
            ticktext=["1880", "1898", "1915", "1933", "1951", "1969", "1986", "2003","2020"],
            showgrid=False, showline=False, showticklabels=True, layer='below traces',
            color="antiquewhite",ticks=""
            ),
    radialaxis = dict(showgrid=True, showline=True,showticklabels=True, linecolor="#3c3b54",gridcolor="#3c3b54",color='antiquewhite',side='clockwise',tickangle=0,type='-',ticks="outside")


    ))
```
